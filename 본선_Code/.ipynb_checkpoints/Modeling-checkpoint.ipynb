```python
import random
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from autogluon.core.metrics import make_scorer
from autogluon.tabular import TabularDataset, TabularPredictor
train = pd.read_parquet('train_v1.parquet')
test = pd.read_parquet('test_v1.parquet')
submit = pd.read_csv('sample_submission.csv')
test_x = test.drop(["ID"],axis=1)
train_data = TabularDataset(train)
test_data = TabularDataset(test_x)
predictor = TabularPredictor(label='target',  eval_metric='mean_absolute_error', problem_type='regression').fit(train_data, presets='best_quality')
y_pred = predictor.predict(test_data)
predictions = predictor.predict(test_data, model='CatBoost_BAG_L1')
y_pred = predictions
pred = pd.DataFrame(y_pred, columns = ['target'])
test_all = pd.concat([test,pred],axis=1)
test_all = test_all[['ID','target']].copy()
sample_submission_df = pd.read_csv('sample_submission.csv')
merged_df = sample_submission_df.merge(test_all, on='ID', how='left')
merged_df = merged_df.drop(columns=['weight'])
merged_df['target'].fillna(400, inplace=True)
merged_df.rename(columns={'target': 'weight'}, inplace=True)
merged_df.to_csv('AG_CAT_RAW.csv',index=False)
predictions = predictor.predict(test_data, model='NeuralNetFastAI_BAG_L1')
y_pred = predictions
pred = pd.DataFrame(y_pred, columns = ['target'])
test_all = pd.concat([test,pred],axis=1)
test_all = test_all[['ID','target']].copy()
sample_submission_df = pd.read_csv('sample_submission.csv')
merged_df = sample_submission_df.merge(test_all, on='ID', how='left')
merged_df = merged_df.drop(columns=['weight'])
merged_df['target'].fillna(400, inplace=True)
merged_df.rename(columns={'target': 'weight'}, inplace=True)
merged_df.to_csv('AG_NNF_RAW.csv',index=False)
predictions = predictor.predict(test_data, model='LightGBMXT_BAG_L1')
y_pred = predictions
pred = pd.DataFrame(y_pred, columns = ['target'])
test_all = pd.concat([test,pred],axis=1)
test_all = test_all[['ID','target']].copy()
sample_submission_df = pd.read_csv('sample_submission.csv')
merged_df = sample_submission_df.merge(test_all, on='ID', how='left')
merged_df = merged_df.drop(columns=['weight'])
merged_df['target'].fillna(400, inplace=True)
merged_df.rename(columns={'target': 'weight'}, inplace=True)
merged_df.to_csv('AG_LGBMXT_RAW.csv',index=False)
```

```python
train = pd.read_parquet('train_v4.parquet')
test = pd.read_parquet('test_v4.parquet')
submit = pd.read_csv('sample_submission.csv')
test_x = test.drop(["ID"],axis=1)
train_data = TabularDataset(train)
test_data = TabularDataset(test_x)
predictor = TabularPredictor(label='target',  eval_metric='mean_absolute_error', problem_type='regression').fit(train_data, presets='best_quality', num_stack_levels=0)
predictions = predictor.predict(test_data, model='CatBoost_BAG_L1')
y_pred = predictions
pred = pd.DataFrame(y_pred, columns = ['target'])
test_all = pd.concat([test,pred],axis=1)
test_all = test_all[['ID','target']].copy()
sample_submission_df = pd.read_csv('sample_submission.csv')
merged_df = sample_submission_df.merge(test_all, on='ID', how='left')
merged_df = merged_df.drop(columns=['weight'])
merged_df['target'].fillna(0, inplace=True)
merged_df.rename(columns={'target': 'weight'}, inplace=True)
merged_df.to_csv('AG8_CAT_RAW.csv',index=False)
predictions = predictor.predict(test_data, model='NeuralNetFastAI_BAG_L1')
y_pred = predictions
pred = pd.DataFrame(y_pred, columns = ['target'])
test_all = pd.concat([test,pred],axis=1)
test_all = test_all[['ID','target']].copy()
sample_submission_df = pd.read_csv('sample_submission.csv')
merged_df = sample_submission_df.merge(test_all, on='ID', how='left')
merged_df = merged_df.drop(columns=['weight'])
merged_df['target'].fillna(0, inplace=True)
merged_df.rename(columns={'target': 'weight'}, inplace=True)
merged_df.to_csv('AG8_NNF_RAW.csv',index=False)
predictions = predictor.predict(test_data, model='LightGBMXT_BAG_L1')
y_pred = predictions
pred = pd.DataFrame(y_pred, columns = ['target'])
test_all = pd.concat([test,pred],axis=1)
test_all = test_all[['ID','target']].copy()
sample_submission_df = pd.read_csv('sample_submission.csv')
merged_df = sample_submission_df.merge(test_all, on='ID', how='left')
merged_df = merged_df.drop(columns=['weight'])
merged_df['target'].fillna(0, inplace=True)
merged_df.rename(columns={'target': 'weight'}, inplace=True)
merged_df.to_csv('AG8_LG8BMXT_RAW.csv',index=False)
a = pd.read_csv('AG_CAT1_RAW.csv')
b = pd.read_csv('AG_NNFL1_RAW.csv')
f = pd.read_csv('AG4_NNFL1_RAW.csv')
h = pd.read_csv('AG4_CATL1_RAW.csv')
k = pd.read_csv('AG7_NNF_RAW.csv')
m = pd.read_csv('AG7_CAT_RAW.csv')
n = pd.read_csv('AG8_NNF_RAW.csv')
p = pd.read_csv('AG8_CAT_RAW.csv')
a['weight'] = a['weight'].apply(lambda x: (x // 100) * 100 if x % 100 <= 44 else ((x // 100) + 1) * 100)
b['weight'] = b['weight'].apply(lambda x: (x // 100) * 100 if x % 100 <= 44 else ((x // 100) + 1) * 100)
f['weight'] = f['weight'].apply(lambda x: (x // 100) * 100 if x % 100 <= 44 else ((x // 100) + 1) * 100)
h['weight'] = h['weight'].apply(lambda x: (x // 100) * 100 if x % 100 <= 44 else ((x // 100) + 1) * 100)
k['weight'] = k['weight'].apply(lambda x: (x // 100) * 100 if x % 100 <= 44 else ((x // 100) + 1) * 100)
m['weight'] = m['weight'].apply(lambda x: (x // 100) * 100 if x % 100 <= 44 else ((x // 100) + 1) * 100)
n['weight'] = n['weight'].apply(lambda x: (x // 100) * 100 if x % 100 <= 44 else ((x // 100) + 1) * 100)
p['weight'] = p['weight'].apply(lambda x: (x // 100) * 100 if x % 100 <= 44 else ((x // 100) + 1) * 100)
min_weights = pd.concat([a['weight'], b['weight'], f['weight'], h['weight'], k['weight'], m['weight'], n['weight'], p['weight']], axis=1).min(axis=1)
submit = pd.read_csv('sample_submission.csv')
submit['weight'] = min_weights
submit['weight'][submit['weight'] > 900] = 900
submit['weight'][submit['weight'] < 0] = 0
weight_updates = {100: 102, 400: 402, 600: 602, 700: 702, 800: 802, 900: 902}
submit['weight'] = submit['weight'].replace(weight_updates)
submit.to_csv('SOTA.csv', index=False)
```
