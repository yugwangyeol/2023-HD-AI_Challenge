# v1

```python
import glob
import os
import pandas as pd
import numpy as np
train_files = glob.glob('./train/*.csv')
window_size = 500  
step = 5  

def create_sliding_window_samples(data, window_size, step):
    samples = []
    num_features = data.shape[1]
    for start in range(0, len(data) - window_size + 1, 1):
        window_samples = [data.iloc[start:start+window_size, i].values[::step] for i in range(num_features)]
        sample = np.concatenate(window_samples)
        samples.append(sample)
    return samples

first_df = pd.read_csv(train_files[0])
feature_names = first_df.columns.drop('Time[s]').tolist()

new_feature_names = []
for feature in feature_names:
    for i in range(window_size // step):
        new_feature_names.append(f"{feature}_{i}")

all_samples = pd.DataFrame()
for file in train_files:
    df = pd.read_csv(file)
    df.drop(['Time[s]'], axis=1, inplace=True)
    
    samples = create_sliding_window_samples(df, window_size, step)
    
    samples_df = pd.DataFrame(samples, columns=new_feature_names)
    
    label = int(os.path.basename(file).split('kg')[0])
    
    samples_df['target'] = label
    
    all_samples = pd.concat([all_samples, samples_df], ignore_index=True)

csv_file_path = 'train_v1.parquet'
all_samples.to_parquet(csv_file_path, index=False)
```
