```python
import os
import pandas as pd
import glob


def calculate_derivative(sensor_data):
    return sensor_data.diff()

def classify_load(sensor_data, threshold=0.01):
    derivative = calculate_derivative(sensor_data)
    static_load = sensor_data[derivative.abs() <= threshold]
    dynamic_load = sensor_data[derivative.abs() > threshold]
    return static_load, dynamic_load

def separate_load_cycles(sensor_data):
    derivative = calculate_derivative(sensor_data)
    rising = sensor_data[derivative > 0]
    falling = sensor_data[derivative < 0]
    return rising, falling

def create_feature_engineering(dataframe):
    dataframe['Torque_A'] = calculate_derivative(dataframe['Sensor A']).abs()
    dataframe['Dynamic_Response_A'] = calculate_derivative(dataframe['Torque_A']).abs()

    dataframe['Torque_D'] = calculate_derivative(dataframe['Sensor D']).abs()
    dataframe['Dynamic_Response_D'] = calculate_derivative(dataframe['Torque_D']).abs()

    return dataframe

if not os.path.exists('train_fe4'):
    os.makedirs('train_fe4')

csv_files = glob.glob('train/*.csv')

for file in csv_files:
    data = pd.read_csv(file)
    
    engineered_data = create_feature_engineering(data)
    engineered_data = engineered_data.fillna(engineered_data.mean())
    engineered_data = engineered_data.reset_index(drop=True)
    new_file_name = 'train_fe4/' + os.path.basename(file).replace('.csv', 'FE.csv')
    
    engineered_data.to_csv(new_file_name, index=False)
```
